```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv('diabetes.csv')
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
df['Outcome'].value_counts()
```

```python
X=df.drop(columns='Outcome',axis=1)
Y=df['Outcome']
```

```python
X.head()
```

```python
Y.head()
```

```python
result = df[df['Outcome'] == 1]
result.head()
```

```python
result1 = df[df['Outcome'] == 0]
result1.head()
```

```python
scaler=StandardScaler()
scaler.fit(X)
```

```python
data=scaler.transform(X)
```

```python
print(data)
```

```python
X=data
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, stratify=Y, random_state=2)
```

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
```

```python
model=DecisionTreeClassifier()
model.fit(X_train,Y_train)
```

```python
prediction=model.predict(X_test)
```

```python
acc=accuracy_score(prediction,Y_test)
print(acc)
```

```python
import pickle
filename = 'diabetes_model.pkl'
pickle.dump(model, open(filename, 'wb'))
```

```python
loaded_model = pickle.load(open('diabetes_model.pkl', 'rb'))
pregnancy=int(input("No. of pregnancy: "))
glucose=int(input("GLucose level: "))
bloodpressure=int(input("Blood pressure: "))
skinthickness=int(input("Skin Thickness: "))
insulin=int(input("Insulin: "))
bmi=float(input("BMI: "))
diabetesPedigreeFunction=float(input("Diabetes pedigree function: "))
age=int(input("Age: "))

input_data=(pregnancy,glucose,bloodpressure,skinthickness,insulin,bmi,diabetesPedigreeFunction,age)

nparray=np.asarray(input_data)
reshaped=nparray.reshape(1,-1)

stddata=scaler.transform(reshaped)
prediction=loaded_model.predict(stddata)
if(prediction==1):
  print("person may be diabetic")

else:
  print("person may not be diabetic")
  #DATA SAMPLES
# 6	148	72	35	0	33.6	0.627	50	1  diabetic
# 1	85	66	29	0	26.6	0.351	31	0  non-diabetic
```
