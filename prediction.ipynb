```python
import pickle
import numpy as np


loaded_scaler = pickle.load(open('diabetes_scaler.pkl', 'rb'))

loaded_model = pickle.load(open('diabetes_model.pkl', 'rb'))
pregnancy=int(input("No. of pregnancy: "))
glucose=int(input("GLucose level: "))
bloodpressure=int(input("Blood pressure: "))
skinthickness=int(input("Skin Thickness: "))
insulin=int(input("Insulin: "))
bmi=float(input("BMI: "))
diabetesPedigreeFunction=float(input("Diabetes pedigree function: "))
age=int(input("Age: "))

input_data=(pregnancy,glucose,bloodpressure,skinthickness,insulin,bmi,diabetesPedigreeFunction,age)

nparray=np.asarray(input_data)
reshaped=nparray.reshape(1,-1)

stddata=loaded_scaler.transform(reshaped)
prediction=loaded_model.predict(stddata)
if(prediction==1):
  print("person may be diabetic")

else:
  print("person may not be diabetic")
  #DATA SAMPLES
# 6	148	72	35	0	33.6	0.627	50	1 for may be diabetic
# 1	85	66	29	0	26.6	0.351	31	0 for may not be diabetic
#pregnancy,glucose,bloodpressure,skinthickness,insulin,bmi,diabetesPedigreeFunction,age
```
